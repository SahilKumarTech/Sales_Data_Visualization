```python
import seaborn as sns
import pandas as pd 
data = pd.read_csv(r'C:\Users\Sahil554\Desktop\sahil\diwali.csv', encoding='latin1')

ax = sns.countplot(x = 'Age Group' ,data = data, hue= 'Gender' )
for bars in ax.containers:
    ax.bar_label(bars)


#This code generates a count plot to visualize the distribution of 'Age Group' in the data.
#It creates a plot with 'Age Group' on the x-axis and adds labels to each bar showing the count.
#Finally, it displays the plot.
```

```python
sales_state = data.groupby(['State'], as_index=False) ['Orders'].sum().sort_values(by= 'Orders',ascending=False).head(9)
sns.set(rc={'figure.figsize':(15,5)})

sns.barplot(x='State', y='Orders', data=sales_state, hue = 'State')
plt.show()

#This code analyzes data to show the top 9 states with the most orders. It groups data by state,
#calculates total orders, sorts results, selects top 9 states, and creates a bar plot to visualize this information.
```

```python
# total amount/sales from 10 states
```

```python
sales_state = data.groupby(['State'], as_index=False) ['Amount'].sum().sort_values(by= 'Amount',ascending=False).head(9)
sns.set(rc={'figure.figsize':(15,5)})

sns.barplot(x='State', y='Amount', data=sales_state , hue = 'State')
plt.show()

#This code analyzes data to show the top 9 states with the highest total amount.
#It groups data by state, calculates total amount, sorts results, selects top 9 states, and creates a bar plot to visualize this information.
```

```python
#from above graphs we can see that unexpectedly most of the order are from uttar pradesh, Maharashtra, and karnataka.
```

```python
import pandas as pd
import seaborn as sns

data = pd.read_csv(r'C:\Users\Sahil554\Desktop\sahil\diwali.csv', encoding='latin1')

sales_state = data.groupby(['Occupation'], as_index=False) ['Amount'].sum().sort_values(by= 'Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x='Occupation', y='Amount', data=sales_state, hue='Occupation')

#This code analyzes data to show the top 10 occupations with the highest total amount. 
#It groups data by occupation, calculates total amount, sorts results, selects top 10 groups, and creates a bar plot to visualize this information.
```

```python
# from above graphs we can see that most of the working in IT, Aviation and Healthcare sector. 
```
